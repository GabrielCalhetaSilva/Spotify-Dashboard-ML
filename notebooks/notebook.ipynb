```python
import pandas as pd
import plotly.express as px
```

```python
#Análise dos dados

df = pd.read_csv('../data/top2018.csv')
```

```python
#Gráfico de dispersão que mostra o X = o quanto a música é dançante e Y = o quanto a música é energética/traz emoções 
#E quando passar o mouse no pontinho você pode ver o nome da música

fig = px.scatter(
    df,
    x='danceability',
    y='energy',
    hover_name='name'
)

fig.show()
```

```python
#Histograma que demonstra que a maioria das músicas tem energia média a alta, o que indica preferência por sons mais animados 

fig2 = px.histogram(df, x='energy')
fig2.show()
```

```python
#As músicas mais dançantes são em sua maioria do estilo trap ou hip-hop

top_dance = df.sort_values(by='danceability', ascending=False).head(10)
top_dance
```

```python
#Principais insights: 
#Energy x Loundness (0.73) = Músicas mais energéticas tendem a ser mais altas, que vem da produção e mixagem mais intensa
#Energy x Acousticness (-0.42) = Quanto mais acústica a música, menos a energia, indicando que sons acústicos muitas vezes são mais calmos, até pela escolha instrumental
#Danceability x Valence (0.41) = Músicas mais dançantes tendem a ser mais positivas/felizes/alegres
#Energy x Valence (0.38) = Músicas mais energéticas na maioria das vezes vão transmitir emoções mais positivas
#Speechiness x Danceability (0.22) = Músicas com mais fala como por exemplo raps e traps, tendem a ser mais dançantes(pelo menos nesse top 100)
#Tempo x Danceability (-0.19) = Esse me surpreendeu pois o tempo(BPM/batidas por minuto) tem uma correlação negativa e bem fraca com o quão dançante a música é, ou seja, músicas super rápidas não são necessariamente as mais dançantes

df_numeric = df.select_dtypes(include='number')
df_numeric.corr()
```
